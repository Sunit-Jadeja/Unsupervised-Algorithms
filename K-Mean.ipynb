```python
import numpy as np
import matplotlib.pyplot as pit
import pandas as pd

dataset= pd.read_csv("/Users/sunitjadeja/Desktop/Python/Mall_Customers.csv")
print(dataset)
x=dataset.iloc[:,[3,4]].values
# print(x)
```

```python
#using the elbow method to find the optimal clusters
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
pit.plot(range(1,11), wcss)
pit.title("The Elbow Method")
pit.xlabel("Number of clusters")
pit.ylabel("wcss")
pit.show()
```

```python
#fitting k-means to the dataset
kmeans=KMeans(n_clusters=5, init="k-means++", random_state=42)
y_kmeans=kmeans.fit_predict(x)
```

```python
pit.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1], s=100,c="red", label="Cluster 1")
pit.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1], s=100,c="blue", label="Cluster 2")
pit.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1], s=100,c="green", label="Cluster 3")
pit.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1], s=100,c="orange", label="Cluster 4")
pit.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1], s=100,c="black", label="Cluster 5")
pit.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
pit.title("Clusters of customers")
pit.xlabel("Annual Income($)")
pit.ylabel("Spending Score(1-100)")
pit.legend()
pit.show()
```
